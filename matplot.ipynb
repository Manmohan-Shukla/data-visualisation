```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
url="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/021/299/original/final_vg1_-_final_vg_%281%29.csv?1670840166"

df =pd.read_csv(url)
df
```

```python
x_val=[0,1,2]
y_val=[3,5,9]
plt.plot(x_val,y_val)
```

```python
cat_counts=df['Genre'].value_counts()
cat_counts.head()
```

```python
x_bar=cat_counts.index
y_bar=cat_counts.values
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90,fontsize=10)
plt.xlabel('Genre',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Count of Games by Genre',fontsize=20)

plt.bar(x_bar,y_bar,width=0.5,color='black')
```

```python
sns.countplot(data=df,x='Genre',order=df['Genre'].value_counts().index,color='orange')
```

```python
sales_data=df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
sales_data
region_Sale=sales_data.T.sum(axis=1)
region_Sale
```

```python
plt.pie(region_Sale,labels=region_Sale.index,autopct='%1.1f%%',startangle=90,explode=[0.1,0,0,0])
plt.title('Total Sales by Region',fontsize=20)  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
url="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/021/299/original/final_vg1_-_final_vg_%281%29.csv?1670840166"

df =pd.read_csv(url)
df
```

```python
toppub=df['Publisher'].value_counts().index[:3]
```

```python
topgen=df['Genre'].value_counts().index[:3]
topplat=df['Platform'].value_counts().index[:3]
toppub,topgen,topplat
```

```python
topdat=df.loc[df['Publisher'].isin(toppub) & df['Genre'].isin(topgen) & df['Platform'].isin(topplat)]
topdat
```

dodge count plot

```python
sns.countplot(data=topdat,x='Publisher',hue='Platform')
```

learn pivot table how you can convert df into pivot table

```python
df_stacked_plt=pd.crosstab(index=topdat['Publisher'],columns=topdat['Platform'])
df_stacked_plt
```

```python
df_stacked_plt.plot(kind='bar',stacked=True)
plt.title('Stacked Bar Plot of Publisher and Platform',fontsize=20)
plt.show()
```
